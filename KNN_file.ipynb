```python
import pandas as pd
#opening file
abalone=pd.read_csv("abalone_csv.csv")
abalone.head()
```

```python
abalone=abalone.drop("Sex",axis=1)
```

```python
import numpy as np
X=abalone.drop("Class_number_of_rings",axis=1)
X=X.to_numpy()
X
```

```python
y=abalone['Class_number_of_rings']
y=y.to_numpy()
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model=KNeighborsRegressor(n_neighbors=3)
```

```python
g=knn_model.fit(X_train,y_train)
g
```

```python

pred=g.predict(X_test)
pred
```
